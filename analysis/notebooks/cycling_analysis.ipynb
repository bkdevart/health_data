# Initial Cycling Analysis

## Daily totals

- Sum up miles by `startDate` day
- Calculate differences in time between startDate/endDate, place in new column as seconds
- Calculate daily total seconds

```python
df = pd.read_csv('../../output/cycling.csv', parse_dates=['start_date'])
df.head()
```

```python
df.info()
```

```python
df.plot(x='start_date', y='miles', title='Daily miles cycled')
```

```python
df.tail()
```

```python
df_date_2024 = df[df['start_date'].dt.date >= dt.date(2024, 1, 1)]
df_date_2024.plot(x='start_date', y='miles', 
                  title='2024 Cycling Miles by Day')
```

```python
df_date_2024_07 = df[df['start_date'] >= dt.datetime(2024, 7, 1)]
df_date_2024_07.plot(x='start_date', 
                     y='miles', 
                     title='07 2024 Cycling Miles by Day', 
                     kind='bar')
```
